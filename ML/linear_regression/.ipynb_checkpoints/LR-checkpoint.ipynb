# Linear Regression Demo Code


## sample LR

### theory

**Data**
$$
    DataSet: (x_1, y_1), (x_2, y_2), ..., (x_{10}, y_{10})
$$
**Model:**
$$
    y = \alpha x + \beta
$$
**Loss Function**
$$
    L = \frac{\sum_{i=1}^n{(y_i - y_i^{'})^2}}{2n}
$$
$$
    y_i^{'} = \alpha x_i + \beta
$$
**Optimization**
1. **最小二乘法**
    $$
        \frac{\partial L}{\partial \alpha} = \frac{\sum_{i=1}^n{(y_i - (\alpha x_i + \beta))(-x_i)}}{n} = 0
    $$
    $$
        \frac{\partial L}{\partial \beta} = \frac{\sum_{i=1}^{n}{((\alpha x_i + \beta)- y_i)}}{n} = 0
    $$
    解得：
    $$
        \alpha = \frac{\sum_{i=1}^n{y_ix_i - \beta x_i}}{\sum_{i=1}^n{{x_i}^2}}
    $$
    $$
        \beta = \frac{\sum_{i=1}^n{(y_i - \alpha x_i)}}{n}
    $$
    
2. **梯度下降法**
    $$
        \alpha = \alpha - \frac{\partial L}{\partial \alpha} \times lr
    $$
    $$
        \beta = \beta - \frac{\partial L}{\partial \beta} \times lr
    $$
    $lr$ 是学习率

### code

```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
import time
```

```python
# 使用 sklearn 的 boston dataset, linear features: 5, 12
data = load_boston()
x = data.data[:,5]  # RM
y = data.target[:]  # MEDV     

# normalization
X_SS = StandardScaler()
x = np.squeeze(X_SS.fit_transform(x[:, np.newaxis]))
Y_SS = StandardScaler()
y = np.squeeze(Y_SS.fit_transform(y[:, np.newaxis]))
```

```python
class Simple_LR():
    def __init__(self, seed=42):
        np.random.seed(seed)
        self.alpha = np.random.rand()
        np.random.seed(seed+1)
        self.beta = np.random.rand()
        
    def get_loss(self, x, y):
        loss = np.sum((y - (self.alpha * x + self.beta))**2)/(2*x.shape[0])
        return loss
    
    def predict(self, x):
        return self.alpha * x + self.beta
    
    def least_squares(self, x, y, max_step=100, time_delay=False):
        alpha = self.alpha
        beta= self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('最小二乘法')
            for step in range(max_step):
                alpha = np.sum(x*(y - beta))/np.sum(x**2)
                beta = np.sum(y-alpha*x)/x.shape[0]
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)   
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
                    
    def gradient_descent(self, x, y, lr=0.01, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('梯度下降法')
            for step in range(max_step):
                alpha_grad = np.sum((y-(alpha*x + beta))*(-x))/x.shape[0]
                beta_grad = np.sum((alpha*x+beta)- y)/x.shape[0]
                alpha = alpha- alpha_grad * lr
                beta = beta - beta_grad *lr
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
    
                
def plot(x0, y0, x1, y1, xlabel="RM", ylabel="MEDV", title='' ,legend=[]):
    # x0 y0 is origin data
    # x1, y1 is predict data
    fig = plt.figure(figsize=(16, 12), dpi=100)
    ax = fig.add_subplot(111)

    titleFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 50,
    }
    LabelFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 30,
    }
    tickFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 30,
    }
    legendFont = {
    'family': 'SimHei',
    'weight': 'bold',
    'size': 30,
    }

    ax.scatter(x, y, s=150, marker='o', color='black')
    print(x1.shape)
    markers = ['*', 's', 'o']
    for i in range(x1.shape[0]):
        ax.plot(x1[i, :], y1[i, :], linewidth=5, marker=markers[i], markersize=30, alpha=0.5)

    ax.set_xticklabels([int(i) for i in ax.get_xticks()], LabelFont)
    ax.set_yticklabels([int(i) for i in ax.get_yticks()], LabelFont)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.tick_params(which='both', direction='in', width=2)

    ax.set_ylabel(f'{ylabel}', **LabelFont)
    ax.set_xlabel(f'{xlabel}', **LabelFont)
    ax.legend(legend, prop=legendFont, frameon=True)
    ax.set_title(title, **titleFont)
    fig.show()
                
```

```python
# 使用 最小二乘法 优化
model = Simple_LR()
model.least_squares(x, y, 100)
model2 = Simple_LR()
model2.gradient_descent(x, y, lr=0.001, max_step=100)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x[:,np.newaxis],y)

plot(x, y, np.stack([x, x, x], axis=0), np.stack([model.predict(x), model2.predict(x), lr.predict(x[:,np.newaxis])], axis=0), title='simple lr', legend=['最小二乘法', '梯度下降法', 'sklearn lr', '源数据'])
```

## Multiple LR

### theory

 **Data**
 $$
     DataSet: ([x_{1, 1}, x_{1, 2}, ..., x_{1, m}], y_1), ..., ([x_{n,1}, x_{n,2}, ..., x_{n,m}], y_n)
 $$
 **Model**
 $$
     y = \alpha_1 x_1 + \alpha_2 x_2 + ... + \alpha_m x_m + \beta
 $$
 **Loss Function**
$$
    L = \frac{\sum_{i=1}^n{(y_i - y_i^{'})^2}}{2n}
$$
$$
    y_i^{'} = \alpha_1 x_{i, 1} + \alpha_2 x_{i, 2} + ... + \alpha_m x_{i, m} + \beta
$$
**Optimization**
1. **最小二乘法**
    $$
        \frac{\partial L}{\partial \alpha_j} =  \frac{\sum_{i=1}^n{ (y_i - y_i^{'})(-x_{i,j}) }}{n} = 0
    $$
    $$
        \frac{\partial L}{\partial \beta} = \frac{\sum_{i=1}^n{y_i^{'}-y_i}}{n} = 0
    $$
    解得
    $$
        \alpha_j = \frac{ \sum_{i=1}^{n}{( y_i x_{i,j} - \sum_{k \neq j}^{m}{(\alpha_k x_{i,k} x_{i,j})} - \beta x_{i,j} } )}{\sum_{i=1}^n x_{i,j}^2}
    $$
    $$
        \beta = \frac{ \sum_{i=1}^{n}{(y_i - \sum_{k=1}^{m}{(a_k x_{i, k})})} }{n}
    $$

2. **梯度下降法**
    $$
        \alpha_j = \alpha_j - \frac{\partial L}{\partial \alpha_j} \times lr
    $$
    $$
        \beta = \beta - \frac{\partial L}{\partial \beta} \times lr
    $$
    $lr$ 是学习率

### code

```python
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
import time
```

```python
# 使用 sklearn 的 boston dataset
data = load_boston()
x = np.transpose(data.data) # Features
y = data.target[:]  # MEDV  

# normailize
X_SS = StandardScaler()
x = np.transpose(X_SS.fit_transform(np.transpose(x)))
Y_SS = StandardScaler()
y = np.squeeze(Y_SS.fit_transform(y[:, np.newaxis]))
```

```python
class Multiple_LR():
    def __init__(self, n_feature, seed=42):
        self.n_feature = n_feature
        alpha = []
        for i in range(self.n_feature):
            np.random.seed(seed+i)
            alpha.append(np.random.rand())
        alpha = np.array(alpha)
        self.alpha = alpha[np.newaxis,:]
        np.random.seed(seed+1000)
        self.beta = np.random.rand()

        
    def get_loss(self, x, y):
        loss = np.sum((y - (np.matmul(self.alpha, x) + self.beta))**2)/(2*x.shape[1])
        return loss
    
    def predict(self, x):
        return np.matmul(self.alpha, x) + self.beta
    
    def least_squares(self, x, y, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('最小二乘法')
            for step in range(max_step):
                for j in range(self.n_feature):
                    if j == self.n_feature - 1:
                        alpha[0, j] = np.sum(y*x[j,:] - np.matmul(alpha[:,:j], x[:j,:])*x[j,:] - beta*x[j,:])/np.sum(x[j,:]**2)
                    if j == 0:
                        alpha[0, j] = np.sum(y*x[j,:] -np.matmul(alpha[:,j+1:], x[j+1:,:])*x[j,:] - beta*x[j,:])/np.sum(x[j,:]**2)
                    else:
                        alpha[0, j] = np.sum(y*x[j,:] - (np.matmul(alpha[:,:j], x[:j,:]) + np.matmul(alpha[:,j+1:], x[j+1:,:]))*x[j,:] - beta*x[j,:])/np.sum(x[j,:]**2)
                beta = np.sum(y - np.matmul(alpha, x))/x.shape[1]
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
        
        
    def gradient_descent(self, x, y, lr=0.01, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('梯度下降法')
            for step in range(max_step):
                alpha_grad = []
                for j in range(self.n_feature):
                    alpha_grad.append(np.sum((y - (np.matmul(self.alpha, x) + self.beta))*(-x[j,:]))/x.shape[1])
                    
                beta_grad = np.sum((np.matmul(self.alpha, x) + self.beta)- y)/x.shape[1]
                
                for j in range(self.n_feature):
                    alpha[0,j] = alpha[0,j] - alpha_grad[j] * lr
                beta = beta - beta_grad *lr
                
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
        
```

```python
# 最小二乘法
model = Multiple_LR(n_feature=13)
model.least_squares(x, y, max_step=200)

# 梯度下降法
model2 = Multiple_LR(n_feature=13)
model2.gradient_descent(x, y, lr=0.15, max_step=100)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import SGDRegressor
# sklearn 的 LR 模型
lr = LinearRegression()
lr.fit(np.transpose(x), y)
# sklearn 的 SGD-LR 模型
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(np.transpose(x), y)

# model.get_loss(x, )

print(f'my muliple lr by least squares, loss:\n\t{mean_squared_error(y, np.squeeze(model.predict(x)))}')
print(f'my muliple lr by gradient descent, loss:\n\t{mean_squared_error(y, np.squeeze(model2.predict(x)))}')

print(f'sklearn by LinearRegression, loss: \n\t{mean_squared_error(y, lr.predict(np.transpose(x)))}')
print(f'sklearn by SGDRegressor,loss: \n\t{mean_squared_error(y, sgdr.predict(np.transpose(x)))}')
```

## sample LR wiht L2 norm

### theory

**Data**
$$
    DataSet: (x_1, y_1), (x_2, y_2), ..., (x_{10}, y_{10})
$$
**Model:**
$$
    y = \alpha x + \beta
$$
**Loss Function**
$$
    L = \frac{\sum_{i=1}^n{(y_i - y_i^{'})^2}}{2n} + \lambda\alpha^2
$$
$$
    y_i^{'} = \alpha x_i + \beta
$$
**Optimization**
1. **最小二乘法**
    $$
        \frac{\partial L}{\partial \alpha} = \frac{\sum_{i=1}^n{(y_i - (\alpha x_i + \beta))(-x_i)}}{n} + 2\lambda \alpha = 0
    $$
    $$
        \frac{\partial L}{\partial \beta} = \frac{\sum_{i=1}^{n}{((\alpha x_i + \beta)- y_i)}}{n} = 0
    $$
    解得：
    $$
        \alpha = \frac{\sum_{i=1}^n{(y_ix_i - \beta x_i)}}{2 \lambda n + \sum_{i=1}^n{{x_i}^2}}
    $$
    $$
        \beta = \frac{\sum_{i=1}^n{(y_i - \alpha x_i)}}{n}
    $$
    
2. **梯度下降法**
    $$
        \alpha = \alpha - \frac{\partial L}{\partial \alpha} \times lr
    $$
    $$
        \beta = \beta - \frac{\partial L}{\partial \beta} \times lr
    $$
    $lr$ 是学习率

### code

```python
class Simple_LR_withL2():
    def __init__(self, lambda_=0.001, seed=42):
        np.random.seed(seed)
        self.alpha = np.random.rand()
        np.random.seed(seed+1)
        self.beta = np.random.rand()
        self.lambda_ = lambda_
        
    def get_loss(self, x, y):
        loss = np.sum((y - (self.alpha * x + self.beta))**2)/(2*x.shape[0])
        return loss
    
    def predict(self, x):
        return self.alpha * x + self.beta
    
    def least_squares(self, x, y, max_step=100, time_delay=False):
        alpha = self.alpha
        beta= self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('最小二乘法')
            for step in range(max_step):
                alpha = np.sum(x*(y - beta))/(np.sum(x**2)+2*self.lambda_*x.shape[0])
                beta = np.sum(y-alpha*x)/x.shape[0]
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)   
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
                    
    def gradient_descent(self, x, y, lr=0.01, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('梯度下降法')
            for step in range(max_step):
                alpha_grad = np.sum((y-(alpha*x + beta))*(-x))/x.shape[0]
                beta_grad = np.sum((alpha*x+beta)- y)/x.shape[0]
                alpha = alpha- alpha_grad * lr
                beta = beta - beta_grad *lr
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
    
                
def plot(x0, y0, x1, y1, xlabel="RM", ylabel="MEDV", title='' ,legend=[]):
    # x0 y0 is origin data
    # x1, y1 is predict data
    fig = plt.figure(figsize=(16, 12), dpi=100)
    ax = fig.add_subplot(111)

    titleFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 50,
    }
    LabelFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 30,
    }
    tickFont = {
        'family': 'SimHei',
        'weight': 'bold',
        'fontsize': 30,
    }
    legendFont = {
    'family': 'SimHei',
    'weight': 'bold',
    'size': 30,
    }

    ax.scatter(x, y, s=150, marker='o', color='black')
    print(x1.shape)
    markers = ['*', 's', 'o']
    for i in range(x1.shape[0]):
        ax.plot(x1[i, :], y1[i, :], linewidth=5, marker=markers[i], markersize=30, alpha=0.5)

    ax.set_xticklabels([int(i) for i in ax.get_xticks()], LabelFont)
    ax.set_yticklabels([int(i) for i in ax.get_yticks()], LabelFont)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)
    ax.tick_params(which='both', direction='in', width=2)

    ax.set_ylabel(f'{ylabel}', **LabelFont)
    ax.set_xlabel(f'{xlabel}', **LabelFont)
    ax.legend(legend, prop=legendFont, frameon=True)
    ax.set_title(title, **titleFont)
    fig.show()
                
```

## Multiple LR with L2 norm

### theory

 **Data**
 $$
     DataSet: ([x_{1, 1}, x_{1, 2}, ..., x_{1, m}], y_1), ..., ([x_{n,1}, x_{n,2}, ..., x_{n,m}], y_n)
 $$
 **Model**
 $$
     y = \alpha_1 x_1 + \alpha_2 x_2 + ... + \alpha_m x_m + \beta
 $$
 **Loss Function**
$$
    L = \frac{\sum_{i=1}^n{(y_i - y_i^{'})^2}}{2n} + \lambda \sum_{j=1}^m{\alpha_j^2}
$$
$$
    y_i^{'} = \alpha_1 x_{i, 1} + \alpha_2 x_{i, 2} + ... + \alpha_m x_{i, m} + \beta
$$
    只对 $\alpha$ 进行 L2 正则项约束
**Optimization**
1. **最小二乘法**
    $$
        \frac{\partial L}{\partial \alpha_j} =  \frac{\sum_{i=1}^n{ (y_i - y_i^{'})(-x_{i,j}) }}{n} + 2 \lambda \alpha_j= 0
    $$
    $$
        \frac{\partial L}{\partial \beta} = \frac{\sum_{i=1}^n{y_i^{'}-y_i}}{n} = 0
    $$
    解得
    $$
        \alpha_j = \frac{ \sum_{i=1}^{n}{( y_i x_{i,j} - \sum_{k \neq j}^{m}{(\alpha_k x_{i,k} x_{i,j})} - \beta x_{i,j} } )}{2 \lambda n +\sum_{i=1}^n x_{i,j}^2}
    $$
    $$
        \beta = \frac{ \sum_{i=1}^{n}{(y_i - \sum_{k=1}^{m}{(a_k x_{i, k})})} }{n}
    $$

2. **梯度下降法**
    $$
        \alpha_j = \alpha_j - \frac{\partial L}{\partial \alpha_j} \times lr
    $$
    $$
        \beta = \beta - \frac{\partial L}{\partial \beta} \times lr
    $$
    $lr$ 是学习率

```python
class Multiple_LR_withL2():
    def __init__(self, n_feature, lambda_=0.001, seed=42):
        self.n_feature = n_feature
        alpha = []
        for i in range(self.n_feature):
            np.random.seed(seed+i)
            alpha.append(np.random.rand())
        alpha = np.array(alpha)
        self.alpha = alpha[np.newaxis,:]
        np.random.seed(seed+1000)
        self.beta = np.random.rand()
        self.lambda_ = lambda_

        
    def get_loss(self, x, y):
        loss = np.sum((y - (np.matmul(self.alpha, x) + self.beta))**2)/(2*x.shape[1])
        return loss
    
    def predict(self, x):
        return np.matmul(self.alpha, x) + self.beta
    
    def least_squares(self, x, y, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('最小二乘法')
            for step in range(max_step):
                for j in range(self.n_feature):
                    if j == self.n_feature - 1:
                        alpha[0, j] = np.sum(y*x[j,:] - np.matmul(alpha[:,:j], x[:j,:])*x[j,:] - beta*x[j,:])/(2*self.lambda_*x.shape[1] + np.sum(x[j,:]**2))
                    if j == 0:
                        alpha[0, j] = np.sum(y*x[j,:] -np.matmul(alpha[:,j+1:], x[j+1:,:])*x[j,:] - beta*x[j,:])/(2*self.lambda_*x.shape[1] + np.sum(x[j,:]**2))
                    else:
                        alpha[0, j] = np.sum(y*x[j,:] - (np.matmul(alpha[:,:j], x[:j,:]) + np.matmul(alpha[:,j+1:], x[j+1:,:]))*x[j,:] - beta*x[j,:])/(2*self.lambda_*x.shape[1] + np.sum(x[j,:]**2))
                beta = np.sum(y - np.matmul(alpha, x))/x.shape[1]
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
        
        
    def gradient_descent(self, x, y, lr=0.01, max_step=100, time_delay=False):
        alpha = self.alpha
        beta = self.beta
        with tqdm(total=max_step) as pbar:
            pbar.set_description('梯度下降法')
            for step in range(max_step):
                alpha_grad = []
                for j in range(self.n_feature):
                    alpha_grad.append(np.sum((y - (np.matmul(self.alpha, x) + self.beta))*(-x[j,:]))/x.shape[1])
                    
                beta_grad = np.sum((np.matmul(self.alpha, x) + self.beta)- y)/x.shape[1]
                
                for j in range(self.n_feature):
                    alpha[0,j] = alpha[0,j] - alpha_grad[j] * lr
                beta = beta - beta_grad *lr
                
                self.alpha = alpha
                self.beta = beta
                loss = self.get_loss(x, y)
                pbar.set_postfix(loss=loss, step=step)
                pbar.update(1)
                if time_delay:
                    time.sleep(1)
        
```

```python
# L2 约束 最小二乘法
model = Multiple_LR_withL2(lambda_=0.005, n_feature=13)
model.least_squares(x, y, max_step=200)


from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import SGDRegressor
# sklearn 的 LR 模型
lr = LinearRegression()
lr.fit(np.transpose(x), y)
# sklearn 的 SGD-LR 模型
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(np.transpose(x), y)

# model.get_loss(x, )

print(f'my muliple lr by L2 norm least squares, loss:\n\t{mean_squared_error(y, np.squeeze(model.predict(x)))}')

print(f'sklearn by LinearRegression, loss: \n\t{mean_squared_error(y, lr.predict(np.transpose(x)))}')
print(f'sklearn by SGDRegressor,loss: \n\t{mean_squared_error(y, sgdr.predict(np.transpose(x)))}')
```
